```python
#Câu 1: Xử lý List 
from random import randrange 
 
print("Chương trình xử lý List") 
n=int(input("Nhập số phần tử")) 
lst=[0]*n 
for i in range(n): 
    lst[i]=randrange(-100,100) 
print("List ngẫu nhiên là:") 
print(lst) 
print("Mời bạn thêm số mới:") 
value=int(input()) 
lst.append(value) 
print(lst) 
print("bạn muốn đếm số nào:") 
k=int(input()) 
dem=lst.count(k) 
print(k,"xuất hiện",dem,"lần trong list") 
def CheckPrime(n):
    d=0 
    for i in range(1,n+1): 
        if n%i ==0: 
            d+=1 
    return d==2 
demnt=0 
tongnt=0 
for x in lst: 
    if CheckPrime(x): 
        demnt+=1 
        tongnt+=x 
print("Có",demnt,"số nguyên tố trong list") 
print("Tổng=",tongnt) 
lst.sort() 
print("List sau khi sort:") 
print(lst) 
 
del lst
print("Đã xóa list") 
```

```python
#Câu 2: Xử lý List nhập ngẫu nhiên
from random import randrange 
 
lst=[]
n=int(input("Nhập số phần tử:")) 
for i in range(n): 
    lst.append(randrange(0,100)) 
print("List sau khi tạo ngẫu nhiên là:") 
print(lst) 
x=int(input("Mời bạn chèn thêm số mới")) 
lst.append(x) 
print("List sau khi chèn:") 
print(lst) 
k=int(input("Mời nhập số để xóa")) 
while lst.count(k)>0: 
    lst.remove(k) 
print("List sau khi xóa:") 
print(lst) 
 
def CheckDoiXung(lst): 
    for i in range(len(lst)): 
        if lst[i]!=lst[len(lst)-i-1]: 
            return  False 
    return True 
kt=CheckDoiXung(lst) 
if kt==True: 
    print("List đối xứng") 
else: 
    print("List không đối xứng")
```

```python
#Câu 3: Xử lý List Đa chiều
from random import randrange 
 
 
def TaoMaTran(m,n): 
    D=[] 
    for i in range(m): 
        row=[] 
        for j in range(n): 
            row.append(randrange(100)) 
        D.append(row) 
    return D 
def XuatMaTran(D): 
    for row in D: 
        for element in row: 
            print(element,end='\t') 
        print() 
def LayDong(r): 
    R=D[r] 
    return R 
def XuatList1Chieu(R): 
    for element in R: 
        print(element,end='\t') 
def LayCot(c): 
    C=[] 
    for i in range(len(D)): 
        C.append(D[i][c]) 
    return C 
def MAX(D): 
    max=D[0][0] 
    for i in range(len(D)): 
        for j in range(len(D[i])): 
            if(max<D[i][j]): 
                max=D[i][j] 
    return  max
m=int(input("Nhập số dòng:")) 
n=int(input("Nhập số cột:")) 
D=TaoMaTran(m,n) 
XuatMaTran(D) 
print("------------------------------")
r=int(input("Mời bạn nhập dòng muốn xuất:")) 
print("Dòng",r,"là:")
XuatList1Chieu(LayDong(r)) 
c=int(input("Mời bạn nhập cột muốn xuất:"))
print("\nCột",c,"là:") 
XuatList1Chieu(LayCot(c)) 
max=MAX(D) 

print("\nSố lớn nhất trong ma trận=",max)
```

```python
#Câu 7: Viết chương trình nhập vào một dãy các số theo thứ  tự tăng, nếu nhập sai 
#quy cách thì yêu cầu nhập lại. In dãy số sau khi đã nhập xong 
def nhap_day_so_tang():
    day_so = []
    print("Bắt đầu nhập dãy số (nhập 'q' để dừng).")
    while True:
        gia_tri_nhap = input(f"Nhập số tiếp theo: ")
        
        if gia_tri_nhap.lower() == 'q':
            if not day_so:
                print("Dãy số rỗng.")
            break
            
        try:
            so_moi = int(gia_tri_nhap)
        except ValueError:
            print("Vui lòng nhập số hoặc 'q' để thoát.")
            continue
            
        # Kiểm tra điều kiện tăng
        if not day_so:
            day_so.append(so_moi)
            print("Dãy hiện tại:", day_so)
        elif so_moi >= day_so[-1]:
            day_so.append(so_moi)
            print("Dãy hiện tại:", day_so)
        else:
            print(f"LỖI: Phải nhập số lớn hơn hoặc bằng {day_so[-1]}. Nhập lại.")
            
    return day_so

# --- Chương trình chính ---
print("--- CHƯƠNG TRÌNH NHẬP DÃY SỐ TĂNG ---")
day_da_nhap = nhap_day_so_tang()
print("\nDãy số cuối cùng bạn đã nhập là:")
print(day_da_nhap)
```

```python
#Câu 8: Viết chương trình nhập vào một dãy n số thực M[0], M[1],..., M[n-1], sắp xếp 
#dãy số theo thứ tự giảm dần. Xuất ra dãy số sau khi sắp xếp.
def nhap_n_so_thuc(n):
    M = []
    print(f"Mời bạn nhập {n} số thực:")
    for i in range(n):
        while True:
            try:
                so = float(input(f"Nhập số thứ {i+1}: "))
                M.append(so)
                break
            except ValueError:
                print("Nhập sai, vui lòng nhập lại số thực.")
    return M

# --- Chương trình chính ---
n = int(input("Bạn muốn nhập bao nhiêu số thực: "))
M = nhap_n_so_thuc(n)

print("\nDãy số bạn vừa nhập:")
print(M)

# Sắp xếp giảm dần
M.sort(reverse=True)

print("\nDãy số sau khi sắp xếp giảm dần:")
print(M)
```

```python
#Câu 9: Xử lý mảng
#Hàm check Prime từ câu 1
M = [3, 6, 7, 8, 11, 17, 2, 90, 2, 5, 4, 5, 8]
print(f"Mảng gốc: {M}")

# Khởi tạo các list rỗng
so_le = []
so_chan = []
so_nguyen_to = []
khong_nguyen_to = []

# Duyệt qua mảng M
for x in M:
    # 1. Phân loại chẵn/lẻ
    if x % 2 == 0:
        so_chan.append(x)
    else:
        so_le.append(x)
        
    # 2. Phân loại nguyên tố/không nguyên tố
    if CheckPrime(x):
        so_nguyen_to.append(x)
    else:
        khong_nguyen_to.append(x)

# Xuất kết quả
print("\n--- KẾT QUẢ PHÂN LOẠI ---")

# Dòng 1: Số lẻ
print(f"Dòng 1 (Số lẻ): {so_le}")
print(f"-> Tổng cộng có {len(so_le)} số lẻ.")

# Dòng 2: Số chẵn
print(f"\nDòng 2 (Số chẵn): {so_chan}")
print(f"-> Tổng cộng có {len(so_chan)} số chẵn.")

# Dòng 3: Số nguyên tố
print(f"\nDòng 3 (Số nguyên tố): {so_nguyen_to}")
print(f"-> Tổng cộng có {len(so_nguyen_to)} số nguyên tố.")

# Dòng 4: Không phải số nguyên tố
print(f"\nDòng 4 (Không phải SNT): {khong_nguyen_to}")
print(f"-> Tổng cộng có {len(khong_nguyen_to)} số không phải SNT.")
```

```python
#Câu 10: Xử lý Ma Trận
# Hàm XuatMaTran lấy từ Câu 3
def NhapMaTran(m, n, ten_matrix):
    print(f"\n--- Nhập ma trận {ten_matrix} (kích thước {m}x{n}) ---")
    D = []
    for i in range(m):
        row = []
        for j in range(n):
            val = int(input(f"Nhập {ten_matrix}[{i}][{j}]: "))
            row.append(val)
        D.append(row)
    return D

def CongMaTran(A, B):
    # Kiểm tra kích thước
    m = len(A)
    n = len(A[0])
    if m != len(B) or n != len(B[0]):
        print("Lỗi: Hai ma trận không cùng kích thước, không thể cộng.")
        return None
        
    C = []
    for i in range(m):
        row_C = []
        for j in range(n):
            tong = A[i][j] + B[i][j]
            row_C.append(tong)
        C.append(row_C)
    return C

def HoanVi(A):
    """Hàm tìm ma trận hoán vị (Transpose) của A"""
    m = len(A)      # Số dòng của A
    n = len(A[0])   # Số cột của A
    
    # Ma trận hoán vị AT sẽ có kích thước n x m
    # Tạo ma trận AT chứa toàn số 0
    AT = [[0] * m for _ in range(n)]
    
    for i in range(m):
        for j in range(n):
            # Gán phần tử (i, j) của A
            # vào vị trí (j, i) của AT
            AT[j][i] = A[i][j]
    return AT

# --- Chương trình chính ---
print("Nhập kích thước cho 2 ma trận A và B (phải cùng kích thước):")
m = int(input("Nhập số dòng m: "))
n = int(input("Nhập số cột n: "))

# Nhập ma trận A và B
A = NhapMaTran(m, n, "A")
B = NhapMaTran(m, n, "B")

print("\nMa trận A:")
XuatMaTran(A)
print("\nMa trận B:")
XuatMaTran(B)

# 1. Cộng ma trận
print("\n--- Cộng Ma Trận (A + B) ---")
C = CongMaTran(A, B)
if C:
    print("Kết quả C = A + B:")
    XuatMaTran(C)

# 2. Hoán vị ma trận
print("\n--- Hoán Vị Ma Trận ---")
AT = HoanVi(A)
print("Ma trận hoán vị của A (AT):")
XuatMaTran(AT)

BT = HoanVi(B)
print("\nMa trận hoán vị của B (BT):")
XuatMaTran(BT)
```
