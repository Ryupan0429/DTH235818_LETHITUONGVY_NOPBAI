```python
#Câu 1: Viết hàm để Tính diện tích tam giác
from math import sqrt 
 
print("Chương trình tính diện tích Tam Giác") 
a=float(input("Nhập cạnh a>0:")) 
b=float(input("Nhập cạnh b>0:")) 
c=float(input("Nhập cạnh c>0:")) 
if (a<=0 or b <=0 or c <=0) or (a+b)<=c or (a+c)<=b or b+c<=a: 
    print("Tam giác không hợp lệ") 
else: 
    cv=a+b+c 
    p=cv/2 
    dt=sqrt(p*(p-a)*(p-b)*(p-c)) 
    print("Diện tích =",round(dt,3))
```

```python
#Câu 2: Viết Hàm để chơi Game Đoán Số 
from random import randrange 
while True: 
    somay=randrange(1,101) 
    solandoan=0 
    win=False 
    while solandoan<7: 
        solandoan+=1 
        songuoi=int(input("Máy đoán [1..100], mời bạn đoán:")) 
        print("Bạn đoán lần thứ ",solandoan) 
        if somay==songuoi: 
            print("Chúc mừng bạn đoán đúng, số máy là=",somay) 
            win=True 
            break 
        if somay>songuoi: 
            print("Bạn đoán sai, số máy > số bạn") 
        elif somay<songuoi: 
            print("Bạn đoán sai, số máy < số bạn") 
    if win==False: 
        print("GAME OVER!, số máy =",somay) 
    hoi=input("Tiếp không?") 
    if hoi=="k": 
        break 
print("Cám ơn bạn đã chơi Game!")
```

```python
#Câu 3: Viết Hàm tính BMI
def BMI(height,weight): 
    return weight/(height**2) 
def PhanLoai(bmi): 
    if bmi<18.5: 
        return "Gầy" 
    elif bmi<=24.9: 
        return "Bình thường" 
    elif bmi<=29.9: 
        return "Hơi Béo" 
    elif bmi<=34.9: 
        return "Béo Phì Cấp Độ 1" 
    elif bmi<=39.9: 
        return "Béo Phì Cấp Độ 2" 
    else: 
        return "Béo Phì Cấp độ 3" 
def NguyCoBenh(bmi): 
    if bmi<18.5: 
        return "Thấp" 
    elif bmi<=24.9:
        return "Trung Bình" 
    elif bmi<=29.9: 
        return "Cao" 
    elif bmi<=34.9: 
        return "Cao" 
    elif bmi<=39.9: 
        return "Rất cao" 
    else: 
        return "Nguy Hiểm" 
height=float(input("Nhập vào chiều cao (m):")) 
weight=float(input("Nhập vào cân nặng (kg):"))
bmi=BMI(height,weight) 
print(f"BMI của bạn= {bmi:.2f}") 
print("Phân loại bạn=",PhanLoai(bmi)) 
print("Nguy cơ bệnh của bạn=",NguyCoBenh(bmi))
```

```python
#Câu 4: Viết Hàm tính ROI
def ROI(dt,cp): 
    return (dt-cp)/cp 
def GoiYDauTu(roi): 
    if roi>=0.75: 
        return "Nên đầu tư" 
    else: 
        return "Không nên đầu tư" 
print("Chương trình tính ROI") 
dt=int(input("Nhập Doanh Thu:")) 
cp=int(input("Nhập chi phí:")) 
roi=ROI(dt,cp) 
print("Tỉ Lệ ROI=",roi) 
print("==>",GoiYDauTu(roi))
```

```python
#Câu 5: Viết hàm đệ qui Fibonacci
def fibonacci(n): 
    if n<=2 : 
        return 1 
    return fibonacci(n-1)+fibonacci(n-2) 
 
def listfibo(n): 
    print("Dãy Fibonacci từ 1 tới {n} là: ")
    for i in range(1,n+1): 
        print(fibonacci(i),end='  ') 
 
print("Số Fibonacci tại vị trí thứ 9 là: ",fibonacci(9)) 
listfibo(9) 
```

```python
#Câu 7: Tính và xuất độ dài đoạn AB
from math import sqrt

def Distance(xA,xB,yA,yB):
    AB = sqrt((xB - xA)**2 + (yB - yA)**2)
    return AB

print("=== TÍNH ĐỘ DÀI ĐOẠN AB ===")
xA = float(input("Nhập xA: "))
yA = float(input("Nhập yA: "))
xB = float(input("Nhập xB: "))
yB = float(input("Nhập yB: "))
print(f"Độ dài đoạn AB = {Distance(xA,xB,yA,yB):.2f}")
```

```python
#Câu 9: Viết chương trình tính căn bậc 2 lồng nhau 
def can_bac_2_long_nhau(n):
    ket_qua = sqrt(n)
    for i in range(n-1, 0, -1):
        ket_qua = sqrt(i + ket_qua)
    return ket_qua

print("\n=== TÍNH CĂN BẬC 2 LỒNG NHAU ===")
n = int(input("Nhập n: "))
ket_qua = can_bac_2_long_nhau(n)
print(f"Kết quả: {ket_qua:.6f}")
```

```python
#Câu 12: Hàm oscillate
def oscillate(start, stop):
    if start < 0:
        for i in range(start, 0):
            yield i
            yield -i
    
    if start <= 0 < stop:
        yield 0
        yield 0
    
    if stop > 0:
        start_pos = max(1, start) if start > 0 else 1
        for i in range(start_pos, stop):
            yield i
            yield -i

print("\n=== HÀM OSCILLATE ===")
print("Kết quả oscillate(-3, 5):")
for n in oscillate(-3, 5):
    print(n, end='   ')
print()
```

```python
#Câu 13: Hàm kiểm tra số hoàn thiện, số thịnh vượng
def tinh_tong_uoc_so(n):
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong

def kiem_tra_so_hoan_thien(n):
    tong_uoc = tinh_tong_uoc_so(n)
    return tong_uoc == n

def kiem_tra_so_thinh_vuong(n):
    tong_uoc = tinh_tong_uoc_so(n)
    return tong_uoc > n

print("\n=== KIỂM TRA SỐ HOÀN THIỆN VÀ SỐ THỊNH VƯỢNG ===")
while True:
    n = int(input("Nhập số nguyên dương n: "))
    if(n>0):
        break
    else:
        print("n phải là số nguyên dương!")

print("Câu a)")
if kiem_tra_so_hoan_thien(n):
    print(f"=> {n} là số hoàn thiện")
else:
    print(f"=> {n} không là số hoàn thiện")

print("Câu b)")
if kiem_tra_so_thinh_vuong(n):
    print(f"=> {n} là số thịnh vượng")
else:
    print(f"=> {n} không là số thịnh vượng")

```
