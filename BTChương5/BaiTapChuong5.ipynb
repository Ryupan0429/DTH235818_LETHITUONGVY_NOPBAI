```python
#Câu 1: Kiểm tra chuỗi đối xứng
def CheckDoiXung(s): 
    flag=True 
    for i in range(len(s)): 
        if s[i]!=s[len(s)-i-1]: 
            flag=False 
            break 
    return flag 
def main(): 
    print("Nhập 1 chuỗi:") 
    s=input() 
    if(CheckDoiXung(s)): 
        print("Chuỗi bạn nhập đối xứng") 
    else: 
        print("Chuỗi bạn nhập không đối xứng") 
while True: 
    main() 
    print("Tiếp không bạn?(c/k):") 
    s=input() 
    if s=="k": 
     break 
print("CẢM ƠN BẠN ĐÃ SỬ DỤNG CHƯƠNG TRÌNH!")
```

```python
#Câu 5: Xử lý chuỗi với các hàm cơ bản
def thong_ke_ky_tu(s):
    count_hoa = 0
    count_thuong = 0
    count_so = 0
    count_trang = 0
    count_dacbiet = 0
    count_nguyen_am = 0
    count_phu_am = 0
    
    NGUYEN_AM = "aeiouyáàảãạăắằẳẵặâấầẩẫậéèẻẽẹêếềểễệíìỉĩịóòỏõọôốồổỗộơớờởỡợúùủũụưứừửữựýỳỷỹỵ"
    NGUYEN_AM_HOA = NGUYEN_AM.upper()
    TAT_CA_NGUYEN_AM = NGUYEN_AM + NGUYEN_AM_HOA
    
    # Duyệt qua từng ký tự trong chuỗi
    for char in s:
        if char.isupper():
            count_hoa += 1
        elif char.islower():
            count_thuong += 1
        elif char.isdigit():
            count_so += 1
        elif char.isspace():
            count_trang += 1
        else:
            # Nếu không phải hoa, thường, số, trắng -> là ký tự đặc biệt
            count_dacbiet += 1
            
        # Kiểm tra nguyên âm/phụ âm (chỉ áp dụng cho ký tự chữ)
        if char.isalpha():
            if char in TAT_CA_NGUYEN_AM:
                count_nguyen_am += 1
            else:
                count_phu_am += 1
                
    # In kết quả
    print("\n--- KẾT QUẢ THỐNG KÊ ---")
    print(f"Số chữ IN HOA: {count_hoa}")
    print(f"Số chữ in thường: {count_thuong}")
    print(f"Số chữ là chữ số: {count_so}")
    print(f"Số ký tự đặc biệt: {count_dacbiet}")
    print(f"Số ký tự khoảng trắng: {count_trang}")
    print("-------------------------")
    print(f"Số chữ là Nguyên Âm: {count_nguyen_am}")
    print(f"Số chữ là Phụ âm: {count_phu_am}")

s_input = input("Nhập vào 1 chuỗi để phân tích:\n")
thong_ke_ky_tu(s_input)
```

```python
#Câu 6: Trích lọc số âm trong chuỗi
import re

def NegativeNumberInStrings(s):
    print(f"Chuỗi đầu vào: '{s}'")
    
    # re.findall(pattern, string) sẽ tìm tất cả các chuỗi con khớp với pattern và trả về một danh sách (list)
    matches = re.findall(r"-\d+", s)
    
    if not matches:
        print("Không tìm thấy số nguyên âm nào trong chuỗi.")
        return

    print("Các số nguyên âm tìm được là:")
    for num_str in matches:
        print(num_str)

# --- Chương trình chính ---
# Ví dụ 1
NegativeNumberInStrings("abc-5xyz-12k9l--p")
print("-" * 20)
# Ví dụ 2
NegativeNumberInStrings("Đây là -100 và kia là -25.5 còn đây là 50")
print("-" * 20)
# Ví dụ 3 (không tìm thấy)
NegativeNumberInStrings("abc 123 xyz")
```

```python
#Câu 7: Tối ưu chuỗi danh từ
def ToiUuChuoi(s):
    s_tam = s.strip()
    arr_words = s_tam.split()
    s_toi_uu = " ".join(arr_words)
    return s_toi_uu

def ToiUuChuoiDanhTu(s):
    # 1. Tối ưu khoảng trắng
    s_da_toi_uu = ToiUuChuoi(s)
    
    # 2. Viết hoa chữ cái đầu mỗi từ
    s_danh_tu = s_da_toi_uu.title()
    
    return s_danh_tu

# --- Chương trình chính ---
s_input = """    
  TRần    
duY 
           thAnH   
"""

print(f"Input: \n'{s_input}'")

s_output = ToiUuChuoiDanhTu(s_input)

print(f"\nOutput: \n'{s_output}'")
```
