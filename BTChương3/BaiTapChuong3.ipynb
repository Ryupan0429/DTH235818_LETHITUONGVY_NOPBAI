```python
#Câu 6 Nhập một số n có tối đa 2 chữ số. Hãy cho biết cách đọc ra dạng chữ.  
def doc_so(n):
    chu_so = ["", "một", "hai", "ba", "bốn", "năm", "sáu", "bảy", "tám", "chín"]
    if n==0: 
        return "Không"
    elif n < 10:
        return chu_so[n]
        return "Không"
    elif n < 100:
        chuc = n // 10
        donvi = n % 10
        doc_chuc = "mười" if chuc == 1 else chu_so[chuc] + " mươi"
        if donvi == 0:
            return doc_chuc
        elif donvi == 1 and chuc > 1:
            return doc_chuc + " mốt"
        elif donvi == 5:
            return doc_chuc + " lăm"
        else:
            return doc_chuc + " " + chu_so[donvi]
        
n=-1
while(n<0 or n>99):
    n=int(input("Nhập n: "))
    if(n<0 or n>99): print("Số bạn nhập phải là số nguyên dương và phải có tối đa 2 chữ số. Nhập lại!")
print("Cách đọc số là:"+ doc_so(n))
```

```python
#Câu 8: Nhập vào 2 giá trị a, b và phép toán ‘+’, ‘-’, ‘*’, ‘/’ . Hãy xuất kết quả theo đúng phép toán đã nhập. 
print("--Nhập giá trị a--") 
a=int(input())
print("Giá trị a vừa nhập là:",a)
print("--Nhập giá trị b--")
b=int(input())
print("Giá trị b vừa nhập là:",b)
print("--Nhập phép tính (+,-,*,/)--")
Pheptoan=input()
print("Phép tính vừa nhập là:",Pheptoan)
if Pheptoan=='+':
    print("Kết quả a + b =",a+b)
elif Pheptoan=='-':
    print("Kết quả a - b =",a-b)
elif Pheptoan=='*':
    print("Kết quả a * b =",a*b)
elif Pheptoan=='/':
    if b==0:
        print("Không thể chia cho 0")
    else:
        print("Kết quả a / b =",round(a/b,3))
else:
    print("Phép tính không hợp lệ")
```

```python
#Câu 9: Nhập vào 1 tháng, xuất ra tháng đó thuộc quý mấy trong năm. 

thang = int(input("Nhập vào tháng (1-12): "))
    
if 1 <= thang <= 12:
    quy = (thang - 1) // 3 + 1
    print(f"Tháng {thang} thuộc quý {quy}")
else:
    print("Tháng không hợp lệ! Vui lòng nhập từ 1-12.")
```

```python
#Câu 10: Tính dãy số 
x=int(input("Nhập x:")) 
n=int(input("Nhập N:")) 
s=0 
for i in range(1,n+1): 
    tu=x**i 
    mau=1 
    for j in range(1,i+1): 
        mau=mau*j 
    s=s+(tu/mau) 
print("s({0},{1})={2}".format(x,n,round(s,3)))
```
